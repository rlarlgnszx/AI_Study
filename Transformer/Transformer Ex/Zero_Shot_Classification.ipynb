```python
from transformers import pipeline
```

### Text Classification

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
text = "I am feeling sick today."
labels = ["normal", "happy", "sad"]
```

```python
result = classifier(text, labels)
```

```python
result
```

### Image Classification

```python
model_name = "openai/clip-vit-large-patch14-336"
classifier = pipeline("zero-shot-image-classification", model = model_name)
```

```python
image_to_classify = "iu.webp"
labels_for_classification =  ["zebra",
                              "lion",
                              "dog"]
labels_for_classification1 =  ["zebra",
                              "lion",
                              "dog",
                              "cat"]
labels_for_classification3 =  ["mom", "dad", "son","dauther",'beautiful girl']
scores = classifier(image_to_classify,
                    candidate_labels = labels_for_classification3)
```
